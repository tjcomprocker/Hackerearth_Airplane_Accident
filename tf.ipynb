```python
from __future__ import absolute_import, division, print_function, unicode_literals
import functools

import numpy as np
import tensorflow as tf
import pandas as pd
import keras.backend as K

np.set_printoptions(precision=3, suppress=True)
```

```python
df = pd.read_csv("train.csv")
target = pd.DataFrame(columns=["Severity"])
target["Severity"] = df["Severity"]
df.drop(["Accident_ID","Severity"],axis=1,inplace=True)
target["Severity"] = pd.Categorical(target["Severity"])
target["Severity"] = target["Severity"].cat.codes
```

```python
dataset = tf.data.Dataset.from_tensor_slices((df.values, target.values))
```

```python
def f1_metric(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val

train_dataset = dataset.shuffle(len(df)).batch(1)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=["accuracy"])
model.fit(train_dataset, epochs=5)
```

```python
pred = model.predict(df)
for point in pred:
    print (tf.math.argmax(point))
```
